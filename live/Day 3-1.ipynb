```python
grp = apt.groupby(by = ['구'])['거래금액'].median().sort_values()
```

```python
sns.boxplot(data = apt, x = '구', y = '거래금액', linewidth = 0.5, 
            flierprops = outlier, order = grp.index)
plt.axhline(y = apt['거래금액'].median(), color = 'red', lw = 1.0, ls = '--')
plt.xticks(rotation = 90);
```

```python
grp = apt['구'].value_counts()
grp.head()
```

```python
myPal2 = np.tile('silver', 25)
myPal2[0] = 'red'
myPal2
```

```python
plot = sns.countplot(data = apt, x = '구', order = grp.index, palette = myPal2)
plt.ylim(0, 14000)
plt.xlim(-1, 25)
plt.xticks(rotation = 90)
for i, v in enumerate(grp):
    plt.text(x = i, y = v + 100, s = f'{v:,}', 
             ha = 'center', va = 'bottom', fontsize = 8);
```

```python
figure = plot.get_figure()
```

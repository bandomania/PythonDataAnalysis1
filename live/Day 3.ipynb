```python
# csv 파일 읽기(인코딩 방식: CP949)
price = pd.read_csv(fileName, encoding = 'CP949')
```

```python
# 열 삭제: 지역코드
price = price.drop(columns = ['지역코드'])
```

```python
# 열이름 변경: 시, 구, 동
price.columns
```

```python
price.columns = ['아파트', '시', '구', '동', '지번', '건축년도', 
                 '거래일', '전용면적', '층', '거래금액']
```

```python
price.head()
```

```python
# 열 자료형 변환: 거래금액(정수), 거래일(날짜/시간)
price['거래금액'] = price['거래금액'].str.replace(',', '').astype(np.int64)
price['거래일'] = price['거래일'].astype(np.datetime64)
```

```python
price.dtypes
```

```python
# 필터링: df1(120억 이상), df2(120억 미만 & 65층 이상)
df1 = price[price['거래금액'].ge(1200000)]
df1
```

```python
df2 = price[price['거래금액'].lt(1200000) & price['층'].ge(65)]
df2
```

```python
# 행이름 초기화: df1
df1 = df1.reset_index(drop = True)
df1
```

```python
price.isna().sum()
```

```python
price[price['지번'].isna()]
```

```python
price[price['지번'].isna()]['아파트'].value_counts()
```

```python
# 결측값 원본으로 채움: 서초포레스타2단지(384), 힐스테이트 서초 젠트리스(557)
price.loc[price['아파트'].eq('서초포레스타2단지'), '지번'] = '384'
price.loc[price['아파트'].eq('힐스테이트 서초 젠트리스'), '지번'] = '557'
```

```python
# 결측값 있는 행 삭제: 건축년도
price = price[~price['건축년도'].isna()]
```

```python
price.shape[0]
```

```python
# 건축년도 자료형 정수로 변환
price['건축년도'] = price['건축년도'].astype(np.int64)
```

```python
# 파생변수 생성: 평당금액, 거래금액(억원), 거래년도
price['평당금액'] = price['거래금액'] / price['전용면적'] * 3.3
```

```python
price['거래금액'] = price['거래금액'] / 10000
```

```python
price['거래년도'] = price['거래일'].dt.year
```

```python
price.head()
```

```python
# 파생변수 생성: 경과년수, 거래월, 금액구간, 주소
price['경과년수'] = price['거래년도'] - price['건축년도']
```

```python
price['거래월'] = price['거래일'].dt.strftime('%m월')
```

```python
locs = price['평당금액'].ge(5000)
locs
```

```python
price['금액구분'] = np.where(locs, '5천 이상', '5천 미만')
```

```python
price.columns
```

```python
cols = ['시', '구', '동', '지번']
price[cols]
```

```python
price['주소'] = price[cols].apply(func = lambda x: ' '.join(x), axis = 1)
```

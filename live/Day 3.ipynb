```python
price = pd.read_csv(fileName, encoding = 'UTF-8', parse_dates = ['거래일'])
```

```python
price.head()
```

```python
price['평당금액'] = price['거래금액'] / price['전용면적'] * 3.3
```

```python
price.head()
```

```python
price['평당금액'].round(3)
```

```python
price['거래금액'] = price['거래금액'] / 10000
```

```python
price.head()
```

```python
pd.reset_option('display.precision')
```

```python
price.head()
```

```python
price['거래년도'] = price['거래일'].dt.year
```

```python
price['경과년수'] = price['거래년도'] - price['건축년도']
```

```python
price.head()
```

```python
price['거래월'] = price['거래일'].dt.strftime('%m월')
```

```python
price.head()
```

```python
locs = price['평당금액'].ge(5000)
locs
```

```python
price['금액구간'] = np.where(locs, '5천 이상', '5천 미만')
```

```python
price.head()
```

```python
imsi = price.drop(columns = ['금액구간']).copy()
```

```python
imsi.head()
```

```python
imsi.loc[imsi['평당금액'].ge(5000), '금액구간'] = '5천 이상'
```

```python
imsi.loc[imsi['평당금액'].lt(5000), '금액구간'] = '5천 미만'
```
